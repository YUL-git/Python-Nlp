# 웹 크롤링에 의한 데이터 수집

## 오늘의 핵심강의 : '교보문고 주간 베스트' 크롤링 코드 이해 및 활용

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import pandas as pd

html = urlopen('http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf')
soup = bs(html, "html.parser")

book_page_urls = []
for cover in soup.find_all('div', {'class':'detail'}):
    link = cover.select('a')[0].get('href')
    book_page_urls.append(link)

dic = []
for index, book_page_url in enumerate(book_page_urls):
    html = urlopen(book_page_url)
    soup = bs(html, "html.parser")
    title = soup.find('meta', {'property':'og:title'}).get('content')
    author = soup.find('meta',{'name':'author'}).get('content')
    price = soup.find('meta', {'property':'og:price'}).get('content')
    dic.append({'Title':title,'Author' : author,'Price':price+'원'})

df = pd.DataFrame(dic)
df[4:]
```

## 가.select와 find 사용법 차이

* select는 CSS 선택자 이용
* find는 태그와 속성, 속성값 이용
* find = select_one, find_all = select

```python
# find와 select_one 사용법 비교
from bs4 import BeautifulSoup as bs
#division paragrah 속에 두개의 p를 가지고 와서 예시를 들어 보자
tag = '''
<div class = 'section'>
<p class='start' id='s1'> 시작합니다! </p>
<p class='end' id='s2'> 마지막입니다! </p>
</div>
'''
#bs로 불러와서 soup 객체에 저장했어
soup = bs(tag)

# # find는 태그 이름, 속성으로 특정, find 는 첫번째 하나만 찾아줘
# print(soup.find('p'))
# print(soup.find(class_='end')) #여러개일 경우 첫번째 나오는 것 출력, 하나일 경우 unique한 것을 뽑아옴
# print(soup.find(attrs = {'class':'start'})) #class=''로 찾아 줄 수 있고, attrs(attributes속성) = {k:v}딕션어리 형태로 태그를 찾아 줄 수 있다.
# print(soup.find('p', {'id':'s2'})) #p 태그속에 id: s2인 것을 찾아줘라 

# # select_one는 css로 특정, find와 사용 규칙이 조금 달라
print(soup.select_one('p'))
print(soup.select_one('.start')) #.start 어떤 태그 속에 start라는 값을 가지고 있는 것을 출력 시켜줘라
print(soup.select_one('p.end')) # p태그 속에 end인 것을 찾아줘라
print(soup.select_one('p#s1')) #p 태그 속에 id가 s1인 것을 찾아줘라
print(soup.select_one('#s2')) #아이디가 s2인 것을 찾아줘
```

```python
# find_all과 select 사용법 비교
from bs4 import BeautifulSoup as bs

soup = bs(tag) 

print(soup.find('div').find('p').text) #div을 찾고 이어서 p를 찾아줘라 그리고 .text로 텍스트만 출력해줘라
print(soup.select_one('div > #s2').text) #select_one으로 해서 div id가 s2인 것을 찾아줘라 텍스트만 출력
#find_all은 모든 문서를 찾아줌
for i in soup.find_all('p'):
  print(i.get('id')) #get 하면 class에 value 값을 가져와
for i in soup.select('p') :
  print(i.get('id')) 
#for문을 활용하여 다 p를 다 가져오기, 텍스트만 가져오고 싶다고 하면 i.text로 해주면 되지
for i in soup.find_all('p'):
  print(i.text)
#중간에 있는 값을 가져오고 싶다면, find_all은 모두 가져오기 때문에 for문을 돌려서 가져와 줘야해
for i in soup.find_all('p'):
  print(i.get('class'))
```

```python
#select도 모두 문서를 모두 가져오기 때문에 for문을 돌려서 가져와줘야해
for i in soup.select('p'):
  print(i)
```

# 5.사례 연구 : 교보문고 주간 베스트

```python
# 필요한 라이브러리 선언, urlopen과 bs로 parsing
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import pandas as pd

html = urlopen('http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf')
soup = bs(html, "html.parser")
```

```python
# 주간 베스트에 해당하는 링크 url을 찾아서 리스트에 담기 
#divsion을 찾고 class가 detail인 것을 크롤링 해 올거야
book_page_urls = []
for cover in soup.find_all('div', {'class':'detail'}):
    link = cover.select('a')[0].get('href') # class detail 안에서 a 단락을 고를거야
    book_page_urls.append(link)
book_page_urls
```

```python
# url별로 순회하면서 딕셔너리 데이터로 담아서 list에 담기
dic = []
for index, book_page_url in enumerate(book_page_urls):
    html = urlopen(book_page_url)
    soup = bs(html, "html.parser")
    title = soup.find('meta', {'property':'og:title'}).get('content')
    author = soup.find('meta',{'name':'author'}).get('content')
    price = soup.find('meta', {'property':'og:price'}).get('content')
    # description = soup.find('meta', {'property':'og:description'}).get('content')
    # print(title, author, price+"원")
    dic.append({'Title':title,'Author' : author,'Price':price+'원'})

```
